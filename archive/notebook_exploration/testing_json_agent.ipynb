```python
import os
import json

from langchain.agents import (
    create_json_agent,
    AgentExecutor
)
from langchain.agents.agent_toolkits import JsonToolkit
from langchain.chains import LLMChain
from langchain.llms.openai import OpenAI
from langchain.requests import TextRequestsWrapper
from langchain.tools.json.tool import JsonSpec
```

```python
with open("../../data/dev/cs_courses_objects.json") as f:
    data = json.load(f)
    data = {"courses": data}
    json_spec = JsonSpec(dict_= data, max_value_length=4000)
    json_toolkit = JsonToolkit(spec=json_spec)

    json_agent_executor = create_json_agent(
        llm=OpenAI(temperature=0),
        toolkit=json_toolkit,
        verbose=True
    )
```

```python
json_agent_executor.run("Which CSCI courses are being taught by Prof. Freund in the fall?")
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.experimental.plan_and_execute import PlanAndExecute, load_agent_executor, load_chat_planner
from langchain.llms import OpenAI
from langchain import SerpAPIWrapper
from langchain.agents.tools import Tool
from langchain import LLMMathChain
```

```python
llm = OpenAI(temperature=0)
tools = [
    Tool(
        name = "Course descriptions",
        func=lambda x: "CSCI 331 teaches you about computer security",
        description="useful for when you need to answer questions about what is taught in each course"
    ),
    Tool(
        name="Section information",
        func=lambda x: "CSCI 331 Section 1: MWF, 9-10am",
        description="useful for when you need to answer questions about the specific sections of each course (meeting times, instructors, grading policies, etc.)"
    ),
]
```

```python
model = ChatOpenAI(temperature=0)
```

```python
planner = load_chat_planner(model)
```

```python
executor = load_agent_executor(model, tools, verbose=True)
```

```python
agent = PlanAndExecute(planner=planner, executer=executor, verbose=True)
```

```python
agent.run("Which course is offered in the Fall and meets in the afternoons, and will teach me about computer security?")
```
